```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import numpy as np
```

```python
SRC = 'https://raw.githubusercontent.com/MoH-Malaysia/data-darah-public/main/donations_state.csv'
df = pd.read_csv(SRC)
df.date = pd.to_datetime(df.date)
```

```python
# daily = df['daily']
# state = df['state']
# date =  df['date']
df = df[['date','state','daily']]
df_23 = df[(df.state == 'Malaysia') & (df.date.dt.year.isin([2023]))]
df_23 = df_23.resample('ME', on='date').sum().reset_index()

plt.barh(df_23,width=0.8)
```
